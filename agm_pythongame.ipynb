```python
import random
import openpyxl
import getpass

RUTA_EXCEL = "UCM.xlsx"

def obtener_numero_aleatorio():
    return random.randint(1, 1000)

def solicitar_numero(mensaje):
    return int(getpass.getpass((mensaje)))

def verificar_numero(eleccion, numero_adivinar, intentos):

    if eleccion == numero_adivinar:
        print("A Winner is You!")
        return True

    else:
        if eleccion < numero_adivinar:
            txt = "El número buscado es mayor."
        else:
            txt = "El número buscado es menor."

        print(txt)

        print(f"Te quedan {intentos - 1} oportunidades.")
        return False

def jugar(intentos, modo_doble=False, numero_adivinar=None):

    if modo_doble is False:
        numero_adivinar = obtener_numero_aleatorio()

    resultado_partida = "You Lose"

    while intentos > 0:

        if modo_doble:
            txt = "Jugador 1: Selecciona tu número: "
        else:
            txt = "Selecciona tu número: "

        eleccion = solicitar_numero(txt)

        if verificar_numero(eleccion, numero_adivinar, intentos):
            resultado_partida = "A Winner is You!"
            break

        if modo_doble:
            eleccion = solicitar_numero("Jugador 2: Selecciona tu número: ")
            if verificar_numero(eleccion, numero_adivinar, intentos):
                resultado_partida = "A Winner is You!"
                break

        intentos -= 1

    if intentos == 0:
        print(f"You Lose. El número buscado era {numero_adivinar}.")

    nombre_jugador = input("Introduce tu nombre: ")
    guardar_estadistica(nombre_jugador, resultado_partida)

    return resultado_partida


def dificultad_menu():

    while True:
        print("Menu dificultad\n1. Fácil (20 intentos)\n2. Medio (12 intentos)\n3. Difícil (5 intentos)")
        try:
            opcion = int(input("Elige una opción: "))
            return [20, 12, 5][opcion - 1]
        except (IndexError, ValueError):
            print("Opción de dificultad incorrecta. Le ruego, selecciona una opción válida.")

def guardar_estadistica(nombre_jugador, resultado):

    try:
        excel_document = openpyxl.load_workbook(RUTA_EXCEL)
        plantilla = excel_document["Sheet"]
    except FileNotFoundError:
        excel_document = openpyxl.Workbook()
        plantilla = excel_document.active
        plantilla.append(["Nombre", "Resultado"])

    plantilla.append([nombre_jugador, resultado])
    excel_document.save(RUTA_EXCEL)
    print('Guardado correctamente en el sistema.')

def leer_estadistica():

    try:
        excel_document = openpyxl.load_workbook(RUTA_EXCEL)
        plantilla = excel_document["Sheet"]
        for row in plantilla.iter_rows(values_only=True):
            print(" ".join(map(str, row)))
    except FileNotFoundError:
        print("Aún no existen estadísticas disponibles.")
    except Exception as e:
        print(f"Fallo al leer las estadísticas: {e}")

def main():

    while True:
        print(" Bienvenido al Juego ADIVINA EL NÚMERO de Programación Python ")
        print("Menu\n1. Partida modo solitario\n2. Partida 2 Jugadores\n3. Estadísticas\n4. Salir")
        try:
            opcion = int(input("Elige una opción: "))
            if opcion == 1:
                jugar(dificultad_menu())
            elif opcion == 2:
                numero_adivinar = solicitar_numero("Jugador 1: El número a adivinar debe ser entre 1 al 1000: ")
                jugar(dificultad_menu(), modo_doble=True, numero_adivinar=numero_adivinar)
            elif opcion == 3:
                leer_estadistica()
            elif opcion == 4:
                print("Muchas Gracias Por Jugar. Game Over.")
                break
        except ValueError:
            print("Opción de menu incorrecta. Le ruego, selecciona una opción válida.")

if __name__ == "__main__":
    main()
```
